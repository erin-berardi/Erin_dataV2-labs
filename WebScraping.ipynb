```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://playback.fm/charts/top-100-songs/2021'
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
#soup.prettify()
```

# getting artist name

```python
#myTable > tbody > tr:nth-child(1) > td:nth-child(2) > a
```

```python
soup.select('td:nth-child(2) > a')[0].get_text()
```

```python
title = []

num_iter = len(soup.select('span > a'))
tClist = soup.select('span > a')

for i in range(num_iter):
    title.append(tClist[i].get_text())

title
```

```python
artist = []

num_iter = len(soup.select('td:nth-child(2) > a'))
art_list = soup.select('td:nth-child(2) > a')

for i in range(num_iter):
    artist.append(art_list[i].get_text())
    
artist
```

```python
top_100 = pd.DataFrame({'artist':artist, 'song_title':title})
top_100
```

```python
top_100.dtypes
```

```python
top_100.replace('\\n','',regex=True,inplace=True)
```

```python
top_100['artist']= top_100['artist'].apply(lambda x: x.strip())
     
#df[cols] = df[cols].apply(lambda x: x.str.strip())
```
