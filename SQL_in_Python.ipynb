# Create a Python connection with SQL database and retrieve the results of the following queries as dataframes:

### How many distinct (different) actors' last names are there?
### Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
### Get all films with ARMAGEDDON in the title.
### Get 10 the longest films.
### How many films include Behind the Scenes content?
### Which kind of movies (rating) have a mean duration of more than two hours?
### Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
#How many distinct (different) actors' last names are there?

query ="""SELECT count(distinct last_name)
FROM actor;"""
data1 = pd.read_sql_query(query, engine)
data1
```

```python
#Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
query = """SELECT *,
CASE 
WHEN DATE_FORMAT(rental_date,'%%a') = 'Sat' then 'Weekend'
WHEN DATE_FORMAT(rental_date,'%%a') = 'Sun' then 'Weekend'
ELSE 'Weekday'
END AS 'column_day_type'
FROM rental;"""
data2 = pd.read_sql_query(query, engine)
data2.head()
```

```python
#Get all films with ARMAGEDDON in the title.
query = """SELECT *
FROM sakila.film
WHERE title LIKE ('%%ARMAGEDDON%%');"""
data3 = pd.read_sql_query(query, engine)
data3.head()
```

```python
#Get 10 the longest films
query = """SELECT *
FROM sakila.film
ORDER BY length DESC
LIMIT 10;"""
data4 = pd.read_sql_query(query, engine)
data4.head()
```

```python
#How many films include Behind the Scenes content?
query = """SELECT *
FROM sakila.film
WHERE special_features = 'Behind the Scenes';"""
data5 = pd.read_sql_query(query, engine)
data5.head()
```

```python
#Which kind of movies (rating) have a mean duration of more than two hours?
query = """select rating, round(avg(length),2) as avg_duration from sakila.film
group by rating
having avg_duration > 120
order by avg_duration asc;"""
data6 = pd.read_sql_query(query, engine)
data6.head()
```

```python
#Rank films by length (filter out the rows that have nulls or 0s in length column).
#In your output, only select the columns title, length, and the rank.
query = """SELECT title, length, DENSE_RANK() OVER(ORDER BY length DESC) as 'Rank'
FROM film
WHERE length != 0 AND length IS NOT NULL;"""
data7 = pd.read_sql_query(query, engine)
data7.head()
```
