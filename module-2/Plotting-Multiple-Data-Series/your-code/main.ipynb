```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('/Users/erinberardi/Desktop/GitHub/dataV2-labs/module-2/Plotting-Multiple-Data-Series/data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
data.sort_values(by =['ItemType','RetailSales', 'RetailTransfers', 'WarehouseSales'], inplace = True)
data.plot(x="ItemType", y=["RetailSales", "RetailTransfers","WarehouseSales"])
```

```python
item_type_bar = data[["RetailSales", "RetailTransfers","WarehouseSales"]].groupby(data["ItemType"]).agg("sum")
item_type_bar.plot.bar()
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
supp_sales = data.pivot_table(index = 'Supplier',values = ['RetailSales','RetailTransfers', 'WarehouseSales'],aggfunc='sum')

supp_sales['Total']=supp_sales['RetailSales']+supp_sales['RetailTransfers']+supp_sales['WarehouseSales']

top_ten = supp_sales.pivot_table(index = 'Supplier',values = 'Total')

top_ten.sort_values(by = ['Total'],inplace = True, ascending = False)
top_ten
top_ten.head(10).plot.barh()

```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
data.dtypes
```

```python
data['date']=data['Year'].map(str)+ '-' +data['Month'].map(str)
data['date']=pd.to_datetime(data['date'], format = '%Y-%m').dt.strftime('%Y-%m')
data.head()
```

```python
#over_time = data.pivot_table(index=['date'],values = ["RetailSales", "RetailTransfers","WarehouseSales"])

over_time = data[["RetailSales", "RetailTransfers","WarehouseSales"]].groupby(data['date']).agg("mean")
over_time
```

```python
over_time.plot()
```
